```python
data = pandas.DataFrame(cbsodata.get_data('80416NED'))
data.head()
```

```python
# 
def maand_als_nummer(maand):
    maand_naam = ['januari', 'februari', 'maart', 'april', 'mei', 'juni', 'juli', 'augustus', 'september', 'oktober', 'november', 'december']
    
    if maand in maand_naam:
        maand_nummer = maand_naam.index(maand) + 1
        return maand_nummer 
```

```python
# Create extra colums to add years, months and days
# empty lists
years = []
months = []
days = []

# for every entry append the row and fill in the extra columns
for entry in data.index:
    periods = data['Perioden'][entry].split(' ')
    year = periods[0]
    month = maand_als_nummer(periods[3])
    day = periods[2]
    years.append(year)
    months.append(month)
    days.append(day)

# Create new colums prepopulated with all values
if not 'year' in data:
    data.insert(5, "year", years)
if not 'month' in data:
    data.insert(6, "month", months)
if not 'day' in data:
    data.insert(7, "day", days)
    
```

```python
data
```

```python
benzine = sns.relplot(data=data, x="year", y="BenzineEuro95_1", kind="line", aspect=0.6*3)
diesel = sns.relplot(data=data, x="year", y="Diesel_2", hue='month', kind="line", aspect=0.6*3)
lpg = sns.relplot(data=data, x="year", y="Lpg_3", hue="month", kind="line", aspect=0.6*3)

benzine.set_ylabels("Benzine prijs in euro").set_xlabels("Jaar")
diesel.set_ylabels("Diesel prijs in euro").set_xlabels("Jaar")
lpg.set_ylabels("Lpg prijs in euro").set_xlabels("Jaar")
```
